```python
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go 
from plotly.subplots import make_subplots
```

```python
tlsa =yf.Ticker("TLSA")
tlsa_data= tlsa.history(period="MAX")
```

```python
tlsa_data.reset_index(inplace=True)
tlsa_data.head()
```

```python
import matplotlib.pyplot as plt

def make_graph(data, title):
    plt.figure(figsize=(10, 6))
    plt.plot(data['Date'], data['Close'], label='Tesla Stock Data', marker='o', linestyle='-', color='b')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.legend()
    plt.grid(True)
    plt.show()

# Assuming you have a DataFrame named tesla_stock_data
# Replace this with your actual DataFrame and adjust column names accordingly
tesla_stock_data = pd.DataFrame({'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
                                 'Close': [700, 720, 710]})

# Provide a title for the graph
graph_title = "Tesla Stock Data Plot"

# Call the make_graph function
make_graph(tesla_stock_data, graph_title)
```
